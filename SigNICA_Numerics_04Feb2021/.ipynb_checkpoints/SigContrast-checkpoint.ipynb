# Implementation of SigIC

## Cross-Shuffles and the 'Semi-Lexicographical' Ordering on $[d]^\star$

We provide an implementation of the 'cross-shuffles'

$$\tag{A}\mathcal{W}_{u,v} := \big\{\tilde{u}\sqcup\kern-7mu\sqcup\,\tilde{v} \mid \tilde{u}\in\langle u\rangle^\star\setminus\{\epsilon\}, \ \tilde{v}\in\langle v\rangle^{\star}\setminus\{\epsilon\}\big\}$$

together with a function which reads off the corresp. 'cross-shuffled' signature cumulants from the log-Signature of $X$. 

The idea is to implement two functions, namely:

- A function sha() which takes two strings as an input and returns their shuffle product (list of words) as an output:

```python
def sha(u, v):
    '''input: u,v : words to be shuffled [strings], n: maximal number of summands [int]
       returns: [u\sha v]_n [list of strings], i.e. the 'first' n summands of the shuffle product of u and v.
       NB: This function is only valid over alphabets consisting of single strings (e.g. i = 10 is a word of length 2)!
    ''' 
    if len(u)*len(v)!=0:  #check if one of the factors is the empty word
        shuff_uv = [w + v[-1] for w in sha(u, v[:-1])] + [w + u[-1] for w in sha(u[:-1], v)] 
        return shuff_uv
    else:
        return [u+v]
```

**Example:** $\texttt{sha('212','4')}$ returns $\texttt{['2124', '2142', '2412', '4212']}$, which corresponds to $p \equiv 1\cdot\texttt{2124} + 1\cdot\texttt{2142} + 1\cdot\texttt{2412} + 1\cdot\texttt{4212} \in \mathbb{R}[d]^\star$. 

- A function lextolin() which takes a string as an input and returns the (linearised) ordinal number of its semi-lexicographical position as an output (so e.g.: lextolin(epsilon)=0, lextolin(1)=1, lextolin(12) = 4 provided alph={1,2}):

For $d\in\mathbb{N}$ fixed, let the function $\psi : [d]^\star \rightarrow \mathbb{N}_0$ be defined by sending a word $w\in[d]^\star$ to its ordinal number (i.e. its 'total ranking position') w.r.t. the *'semi-lexicographical'* ordering $\preceq$ on $[d]^\star$ which is given by  

$$\begin{equation}\tag{1}
w_1 \ \prec \ w_2 \qquad :\Longleftrightarrow\qquad \Big\{\big(|w_1| < |w_2|\big)\quad\text{or}\quad \big(|w_1| = |w_2| \quad \text{and}\quad w_1 <_{\mathrm{lexic} }w_2\big)\Big\}
\end{equation}$$

(i.e., $\preceq$ orders two words in $[d]^\star$ by their word length ('the smaller the lower') or, where both of these lenghts are equal, by their lexicographical order ('the earlier the lower'); for instance, $444 \prec 1111$ and $1344 \prec 1444$). The above 'ranking function' $\psi$ then reads

$$\begin{equation}\tag{2}
\psi(\texttt{i}_1\cdots\texttt{i}_k) = \sum_{j=1}^k i_j\cdot d^{k-j} \qquad (i_\nu\in[d], \ k\in\mathbb{N})
\end{equation}$$

(for $\epsilon$ the empty word, we set $\psi(\epsilon):=0$).

*Proof:* We proceed by induction on the length $k:=|w|$ of a given word $w\equiv i_1\cdots i_k\in[d]^\star$. For $k=1$, i.e. $w=\texttt{i}$ for some $i\in[d]$, we clearly have $\psi(w) = i$ as claimed. Let next $w = \tilde{w}\texttt{i}$ for any letter $i\in[d]$ and subword $\tilde{w}\equiv i_1\cdots i_k\in[d]^\star$ of length $|\tilde{w}| = k\in\mathbb{N}$. Then clearly $\psi(w) = d\cdot\psi(\tilde{w}) + i $ (by definition of the ordering (1) on $[d]^\star$) and hence 

$$\begin{equation}\tag{3} \psi(w) = d\cdot\psi(\tilde{w}) + i \stackrel{\mathrm{IH}}{=} d\cdot\left(\sum_{j=1}^k i_j\cdot d^{k-j}\right) + i = \sum_{j=1}^k i_j\cdot d^{k+1-j} + i = \sum_{j=1}^{k+1}i_j\cdot d^{k+1-j} \qquad(\text{for } \ i_{k+1}:= i).
\end{equation}$$

We define $\texttt{lextolin}(w,d) := \psi(w)$ and implement it below:

```python
def lextolin(w, d):
    '''input: word w \in [9]* [string of characters from {1,...,d}], dimension d [=1,...,9].
       returns: the ordinal number [int] of w w.r.t. the 'semi-lexicographical' order (1) on [d]*.
    '''
    k = len(w)
    if k==0:
        return 0
    else:
        a = np.asarray(list(map(int, list(w)))) #convert word to list of its letters (preserving their word-order)
        b = np.asarray([d**(k-(j+1)) for j in range(k)])
        return np.sum(a*b)  # = psi(w) (according to formula (2) above)
```

We now implement a function which returns all cross-shuffles of a given pair of words $(u,v)\in([d]^\star)^{\times 2}$ up to a given length $k$, that is all elements of the set

$$\tag{4}\mathcal{W}_k(u,v) := \mathcal{W}_{u,v}\cap(\mathbb{R}[d])_{\leq k}\qquad (k\in\mathbb{N}).$$

To this end, we implement the following auxiliary functions:
- A function $\texttt{tracespring(u,l)}$ which for $u\in[d]^\star$ and $\ell\in\mathbb{N}$ returns all elements of $\langle u\rangle^\star\cap[d]^\star_\ell$:  

```python
def tracespring(u,l):
    '''input: word u whose trace is of interest [string with chars '1',...,'9'], 
              maximal length l of offspring words [int]
       returns: a list of all words in <u>* which are of length l'''
    
    tru       = set(u)       # determine trace of u
    tru       = ''.join(tru) # join all letters in the trace of u to one string
    truStar_l = [''.join(x) for x in itertools.product(tru, repeat=l)] # generate all length-l-words in <u>*
    
    return truStar_l    
```

**Example:** $\texttt{tracespring('1232',2)}$ returns $\texttt{['22', '21', '23', '12', '11', '13', '32', '31', '33']}$, which corresponds to the family of 2-letter words $\langle \texttt{1232}\rangle^\star\cap[d]^\star_2 = \{\texttt{22},\texttt{21},\texttt{23},\texttt{12},\texttt{11},\texttt{13},\texttt{32},\texttt{31},\texttt{33}\}\subset [d]^\star$ (for any $d\geq 3$). 

- A function $\texttt{crosscumu(u,v,k)}$ which for $u,v\in[d]^\star$ and $k\in\mathbb{N}$ returns those elements of $\mathcal{W}_{u,v}$ which are of degree $k$ (all elements of $\mathcal{W}_{u,v}$ are 1-coefficiened homogeneous word-polynomials, i.e. (free) sums of words of the same length):

```python
def crosscumu(u,v,k):
    '''returns: a list of all (homogeneous; there are not others) degree-k polynomials 
                contained in $\W_{u,v}$ [list of lists]'''
    
    W_k_uv    = [] #list of all desired word-polynomials
    # 'i + j = k'
    for i in range(1,k): #go over all 1\leq i \leq k-1, with i = |\tilde{u}| (cf. (A))
        
        u_til = tracespring(u,i)     # generate \tilde{u} and 
        v_til = tracespring(v, k-i)  # \tilde{v} as in (A)
        
        for ut in u_til:
            for vt in v_til:
                q = sha(ut,vt)       # generate u_t\shuffle v_t (word-polynomial => ~list of strings)
                W_k_uv.append(q)
        
    return W_k_uv
```

**Example:** $\texttt{crosscumu('11','2',3)}$ returns $\texttt{[['122', '212', '221'], ['112', '121', '211']]}$, which corresponds to the family of degree-3 (homogeneous) word polynomials 
$\{q_1\equiv1\cdot\texttt{122} + 1\cdot\texttt{212} + 1\cdot\texttt{221}, q_2\equiv1\cdot\texttt{112} + 1\cdot\texttt{121} + 1\cdot\texttt{211}\}\subset\mathcal{W}_{u,v}$. 

## SigIC-Contrast

Next we finally implement the *SigIC*-contrast function given by 

$$\begin{equation}\tag{5}
\Phi \ : \ X \ \mapsto \ \bar{\kappa}_X := \sum_{k=2}^d\sum_{q\in\mathcal{W}_{{u_k}, {v_k}}}(\kappa^X_q)^2 \qquad \text{ for } \quad u_k\equiv\texttt{12}\cdots\texttt{(k-1)}\quad\text{and}\quad v_k\equiv\texttt{k}.
\end{equation}$$

In place of the *full cumulant series* (5) (which, a priori, may have infinitely many non-zero terms (but will converge under suitable assumptions on $X$)), we for a given tuple $\mu\equiv (m_2, \cdots, m_d)\in\mathbb{N}_0^{d-1}$ in fact compute and consider only the approximation of $\Phi$ which is given by the truncated cumulant series

$$\begin{equation}\tag{6}
\Phi_\mu \ : \ X \ \mapsto \ \bar{\kappa}_X^{(m)} := \sum_{k=2}^d\sum_{q\in\mathcal{W}_{m_k}\!({u_k}, {v_k})}(\kappa^X_q)^2 \qquad \text{ for } \quad \mathcal{W}_{m_k}\!({u_k}, {v_k}) \ \text{ as defined in } \  (4).
\end{equation}$$

To this end, we first implement a function $\texttt{sigcumulextract(X, W)}$ which for an $\mathbb{R}^d$-valued time series $X$ ($\sim$ observation matrix of dimension $n\times d$, with $n$ the total number of observations) and a list $\mathcal{W}$ of 1-coefficiened word-polynomials $q\equiv\sum_w w\in\mathbb{R}[d]$ ($\sim$ list of strings) extracts the cumulant coefficient  

$$\tag{7}\kappa^X_q \ := \ \langle \log[\mathbb{E}\mathfrak{S}(X)], \, q\rangle \ \equiv \ \sum_{w\in q}\langle \log[\mathbb{E}\mathfrak{S}(X)],\,w\rangle.$$

- The function $\texttt{sigcumulextract()}$, in turn, is built on top of the function $\texttt{ESig(X,m,n)}$ which, for $m\in\mathbb{N}$ and $X$ a given time series (pytorch tensor of shape (N,d), $N\in\mathbb{N}$ the number of total observations), returns the empirical estimate of the expected signature of $X$ up to and including its $k^{\mathrm{th}}$ tensor-level (computed upon dividing the totality of $N$ fixed-time observations into $\lfloor\frac{N}{n}\rfloor$-many 'path-snipplets' of window length $n$), i.e.:

$$\begin{equation}\tag{8}
\texttt{ESig(X,m,n)} \ \equiv \ \sum_{j=1}^{N'}\frac{1}{N'}\sum_{w\in[d]^\star_{\leq m}}\langle \mathfrak{S}(\hat{X}_{j}),\,w\rangle\cdot w \ \approx \ \pi_{[m]}\!\left(\mathbb{E}\mathfrak{S}(X)\right) \qquad \text{ for }\quad N':=\lfloor\frac{N}{n}\rfloor
\end{equation}$$

  and $\hat{X}_j$ the piecewise-linear interpolation of the data-batch $(X_{(j-1)n+1}, X_{(j-1)n+2}, \cdots,   X_{jn})\in\mathbb{R}^{d\times n}$. 
  
  **Importantly**, in accordance with the 'semi-lexicographical order' (1) the estimate (8) is returned as a one-dimensional pytorch-tensor of shape $d_m:=\sum_{\nu=1}^m d^\nu$ whose $j^{\mathrm{th}}$-entry corresponds to (the estimate of) the coefficient $\langle \mathbb{E}\mathfrak{S}(X), w_j\rangle$ for $w_j:=\psi^{-1}(j)\in[d]^\star$ with $\psi$ as in (2), i.e. the output vector $t\equiv(t_1, t_2, \cdots, t_{d_m})$ of $\texttt{ESig(X,m,n)}$ is given as    

  $$\begin{equation}\tag{9}
t_{\psi(w)} \ \hat{=} \ \langle \mathbb{E}\mathfrak{S}(X), w\rangle \quad \text{ for all } \ w\in[d]^\star_{\leq m}.
\end{equation}$$

```python
def ESig(X,m,winlen):
    '''input: data X [pytorch tensor of shape (N,d)], tensor level m [int], windowlength of each sample-ts [winlen; int]
       returns: pytorch tensor of shape (N',) containing the values of (8) according to the indexation (9).
    '''
    Y     = torch.clone(X) # copy data
    [N,d] = Y.shape        # collect dimensions of data
    Np    = N//winlen      # N' = \lfloor(N/n)\rfloor
    Y     = Y[:Np*winlen]  # cut data so that it can be 'divided into obvervational batches without remainder'
    
    # Split Y into Np-many chunks of length winlen:
    Ys    = torch.split(Y, winlen) # Ys is a tuple of length Np containing tensors of shape (winlen, d)
    Ys    = torch.stack(Ys)        # Ys is tensor of shape (Np,winlen,d)
    
    # Compute signature (up to depth m) of each \hat{X}_j, j=1,...,Np:
    SigX  = signatory.signature(Ys, m) # tensor of shape (Np, d_m) (with d_m as defined in the text); 
                                       # 'jth-row' of SigX is signature (up to depth m) of batch \hat{X}_j, indexed ~ (9)
    
    # Compute arithmetic mean of rows of SigX (~ (8)):
    ESigX = torch.mean(SigX.double(), 0) # (Convert SigX into a tensor of double-entries so that mean can be computed.)
    
    return ESigX
```

With the (canonical) estimates of (9) at hand, we can now implement $\texttt{sigcumulextract()}$. 

- For this, we implement the auxiliary function $\texttt{coeffextract(t,q,d)}$ which for a given tensor $\mathfrak{t}\in\mathbb{R}^d_{[k]}:=\prod_{m=0}^k(\mathbb{R}^d)^{\otimes k}$ and $q\in\mathbb{R}[d]_{\leq k}$ returns the coefficient $\langle \mathfrak{t},\, q\rangle$. (Here as above, we follow the 'semi-lexicographical order' (1) and encode the tensor $\mathfrak{t}$ as a (1d-)pytorch vector of dimension $d_k$.)

```python
def deg_poly(q):
    '''input: word polynomial q [list of strings]
       returns: the degree of q (ie, length of the longest word in q) [int]
    '''
    return len(max(q, key=len)) # returns length of the longest string in q

def m_dm(dm,d):
    '''Computes number of summands m [int] from the value dm of the geometric sum d_m defined above.'''
    m = np.log(dm*(d-1) + d)/np.log(d) - 1 #the formula for the inversion d_m --> m follows from the geometric sum.
    return int(round(m))

def coeffextract(t,q,d):
    '''input: tensor in $\mathbb{R}^d_{[k]}$, indexed according to (9) [pytorch tensor of shape (N',)], 
              'coefficient polynomial' q [list of strings], d [dimension of X; int] 
       returns: coefficient <t,q> (cf. (7)) [double]. 
    '''
    # Check whether depth of t is at least as large as deg(q):
    d_ESX     = m_dm(len(t), d) # depth of esig_X
    m         = deg_poly(q)     # deg of q
    if m > d_ESX:
        print('The provided tensor is of insufficient depth.')
        return
    
    # Compute c:=<t,q>, using that <t,q> = t_{\psi(q)} (cf. (9), for \psi the linear extension of (2))
    c         = 0
    for w in q:
        c     += t[lextolin(w,d)-1] #gives zero-dimensional torch tensor; coefficients are retrieved according to (9)
                                                                          # (subtract 1 because of Python-indexation)
    return c.item()   #returns the numerical value of c  
```

Prior to the below implementation of $\texttt{sigcumulextract()}$, we recall the following

- **Fact**: The truncated image $G_{[k]}:=\mathrm{range}(\pi_{[k]}\circ\mathfrak{S})\equiv\pi_{[k]}(\mathfrak{S}(\mathrm{BV}))$ is a (nonlinear) submanifold of $\mathbb{R}^d_{[k]}:=\prod_{m=0}^k(\mathbb{R}^d)^{\otimes k}$. As the monomials $\{e_{i_1}\otimes\cdots\otimes e_{i_m} \mid i_j\in[d], \ m\in[k]_0\}\cong[d]^\star_{[k]}$ are the canonical basis of the *linear* space $\mathbb{R}^d_{[k]}$, the most parsimonious coordinate description of $G_{[k]}$ w.r.t. this basis $[d]^\star_{[k]}$ can be achieved after first *linearising* the nonlinear subspace $G_{[k]}$ via a suitable change of coordinates $\phi : \mathbb{R}^d_{[k]}\rightarrow\mathbb{R}^d_{[k]}$. 

  A diffeomorphism $\phi$ which achieves such a linearisation is the *logarithm* $\log : \mathbb{R}^d_{[k]}\rightarrow\mathbb{R}^d_{[k]}$. The *affine subspace* $\log(G_{[k]})\subset\mathbb{R}^d_{[k]}$ will be called the *logarithmic coordinates of the truncated signatures up to depth $k$*.

```python
def sigcumulextract(X,winlen,W): # (this function is needed for (meta) detail-checks only)
    '''input: data X [pytorch tensor of shape (N,d)], windowlength of sample-ts [winlen; int],
              coefficient-polynomials W [list of list of strings]
       returns: signature cumulants (7) for all q in W [list (of length |W|) of doubles].
    '''
    # Determine largest degree of elements in W:
    m         = max([deg_poly(q) for q in W])
    
    # Compute Expected Signature of X up to depth m: 
    esig_X    = ESig(X,m,winlen)
    esig_X    = esig_X.unsqueeze(0) #signatory needs the tensor esig_X to be 2-dimensional, dims corr. to (batch,channels)
    
    # Compute logarithmic coordinates of ESigX (wrt monomial standard basis of \mathbb{R}^d_{[k]}):
    d         = X.shape[1] #collect number of channels of X
    logEsig_X = signatory.signature_to_logsignature(esig_X, d, m, mode='expand') #--> torch.tensor of shape(N',), ind.~(9)
    logEsig_X = logEsig_X.squeeze(0) #reduce superfluous first dimension (which was just needed for signatory())
    
    # Extract the desired coordinates {<logES(x), q> | q in W}:
    sigcums   = []
    for q in W:
        sigcums.append(coeffextract(logEsig_X,q,d))
    
    return sigcums

def logESig(X,m,winlen):
    '''input: data X [pytorch tensor of shape (N,d)], windowlength of sample-ts [winlen; int], tensor depth m [int]
       return: log-coordinates of ESig(X) up to depth m [pytorch tensor of shape (N',), indexed ~ (9)]
    '''
    # Compute Expected Signature of X up to depth m:
    esig_X    = ESig(X,m,winlen)
    esig_X    = esig_X.unsqueeze(0) #signatory needs the tensor esig_X to be 2-dimensional, dims corr. to (batch,channels)
    
    # Compute logarithmic coordinates of ESigX (wrt monomial standard basis of \mathbb{R}^d_{[k]}):
    d         = X.shape[1] #collect number of channels of X
    logEsig_X = signatory.signature_to_logsignature(esig_X, d, m, mode='expand') #--> torch.tensor of shape(N',), ind.~(9)
    
    return logEsig_X.squeeze(0) #reduce superfluous first dimension (which was just needed for signatory())
```

Given the above auxiliary functions, we can now implement the (approximative) contrast function $\Phi_\mu$ introduced in (6):

```python
def ret_uv(k):
    '''input: k [int]
       returns: the words u ='12...(k-1)' and v='k' [strings].''' # Example: ret_uv(5) returns ('1234', '5')
    u = ''
    v = str(k)
    j = 1
    while len(u)<(k-1):
        u += str(j)
        j += 1
    return u,v

def SigIC(X,mu,winlen):
    '''input: data X [pytorch tensor of shape (N,d)], windowlength of sample-ts [winlen; int] (cf. (8)),
              'control-tuple' mu [np-array of shape (d-1,); entries ordered by number of channel] 
       returns: Phi_mu(X) (sum over mu-controlled subselection of cross-cumulants (6)) [double] 
    '''
    d      = X.shape[1] #number of components of X
    
    # Create lists W={W_k, k=2,...,d} of cross-polynomials (cf. (6)/(4)) whose assoc. cumulants (7)/(6) we want to compute:
    W      = [[]]*(d-1)
    mu_max = np.zeros(d-1) #allocate to determine length of longest monomial in W
    
    for k in range(d-1):
        u,v       = ret_uv(k+2)                   #set u=u_k = '12...(k-1)', v=v_k = 'k' for k=2,...,d
        for j in range(int(round(mu[k]))):
            W[k] = W[k] + crosscumu(u,v,j+1)      #assemble W_k(u_k,v_k) = \sqcup_{j=1,...,mu[k]} crosscumu(u,v,j) (cf. (4))
        mu_max[k] = max([deg_poly(q) for q in W[k]]) #determine length of longest monomial in W[k]
        
    mu_max = int(round(np.max(mu_max))) #length of longest monomial in W
    
    ### Print W if you want to see over which cross-shuffled elements q the sum (6) is taken:
    print('W =', W) 
    
    # Compute log-coordinates of ESig(X) up to depth mu_max
    logESX = logESig(X,mu_max,winlen)
    
    s      = np.zeros(d-1)
    # Compute the kth-summand of the outer sum in (6) 
    for k in range(d-1):
        s[k] = sum([coeffextract(logESX,q,d)**2 for q in W[k]]) #k-th inner sum in (6) [cf. (9)]
    
    # Compute \Phi_mu(X) = \sum_{k=0}^{d-2} s[k] (cf. (6))
    SigICX  = np.sum(s)
    
    return SigICX
```

## Add Scaling and Outlier Removal (modify SigIC to make it employable as an ICA-CF)

It is clear that in its given form (5) (resp. (6)), the function $\Phi$ depends on the amplitude of its argument-channels and is hence not yet suitable as a Contrast Function for ICA. In order to achieve scale-invariance as desired (and thus turn $\texttt{SigIC}$ into a proper 'correlation measure'), we propose to *normalize* $\Phi$ as follows: For a given word $w\equiv \texttt{i}_1\cdots \texttt{i}_k\in[d]^\star_k$, let $m_i:=\sum_{\nu\in[k]}\delta_{i, i_k}$ denote the number of times the letter $\texttt{i}$ appears in $w$. We introduce the following normalisation:

$$\begin{equation}\tag{12}
\text{for } \quad \kappa_{w} \quad\text{define}\quad \bar{\kappa}_{w}:= \frac{\kappa_w}{\tilde{\kappa}_{1}^{m_1}\cdot\ldots\cdot\tilde{\kappa}_{d}^{m_d}} \quad\text{for}\quad\tilde{\kappa}_{i}:=\kappa_{ii}^{1/2}.
\end{equation}$$

We normalise $\Phi_\mu$ accordingly by setting

$$\begin{equation}\tag{13}
\bar{\Phi}_\mu \ : \ X \ \mapsto \ \sum_{k=2}^d\sum_{q\in\mathcal{W}_{m_k}\!({u_k}, {v_k})}(\bar{\kappa}^X_q)^2 \qquad\text{with}\qquad\bar{\kappa}^X_q:= \sum_{w\in q}\bar{\kappa}_w.
\end{equation}$$

An implementation of $\bar{\Phi}_\mu$ is given below.

```python
def barSigCF(X,mu,winlen):
    '''input: data X [pytorch tensor of shape (N,d)], windowlength of sample-ts [winlen; int] (cf. (8)),
              'control-tuple' mu [np-array of shape (d-1,); entries ordered by number of channel] 
       returns: \bar{Phi}_mu(X) (sum over mu-controlled subselection of normalized cross-cumulants (13)) [double] 
    '''
    d      = X.shape[1] #number of components of X
    
    # Create lists W={W_k, k=2,...,d} of cross-polynomials (cf. (6)/(4)) whose assoc. cumulants (7)/(6) we want to compute:
    W      = [[]]*(d-1)
    mu_max = np.zeros(d-1) #allocate to determine length of longest monomial in W
    
    for k in range(d-1):
        u,v       = ret_uv(k+2)                   #set u=u_k = '12...(k-1)', v=v_k = 'k' for k=2,...,d
        for j in range(int(round(mu[k]))):
            W[k] = W[k] + crosscumu(u,v,j+1)      #assemble W_k(u_k,v_k) = \sqcup_{j=1,...,mu[k]} crosscumu(u,v,j) (cf. (4))
        mu_max[k] = max([deg_poly(q) for q in W[k]]) #determine length of longest monomial in W[k]
        
    mu_max = int(round(np.max(mu_max))) #length of longest monomial in W
    
    ### Print W if you want to see over which cross-shuffled elements q the sum (6) is taken:
    print('W =', W) 
    
    # Compute log-coordinates of ESig(X) up to depth mu_max
    logESX = logESig(X,mu_max,winlen)
    
    s      = torch.zeros(d-1)
    # Compute the kth-summand of the outer sum in (13):
    for k in range(d-1):
        # Compute normalization (12) for each q in W[k]:
        norms = dict.fromkeys([tuple(q) for q in W[k]]) #allocate dictionary (with q \in W[k] as keys, i.e.: norms = {q: None}) 
        for q in W[k]:
            ordrep          = ''.join(sorted(q[0])) #returns ordered representative \vec{q} of q
            normfactors_0   = [''.join(g) for _, g in groupby(ordrep)] #select 'monoletteral subwords' of \vec{q}
            # Compute factors according to (12):
            normfactors_1   = torch.tensor([coeffextract(logESX,[f[0]+f[0]],d)**(-len(f)/2) for f in normfactors_0])
            norms[tuple(q)] = torch.prod(normfactors_1) #compute normalisation constant (~ denominator in (12))
        s[k] = torch.sum(torch.tensor([(norms[tuple(q)]*coeffextract(logESX,q,d))**2 for q in W[k]])) #k-th inner sum in (13)
    
    # Compute \Phi_mu(X) = \sum_{k=0}^{d-2} s[k] (cf. (6))
    SigICX  = torch.sum(s)
    
    return SigICX
```

As an alternative approach to achieve scale-invariance of $\texttt{SigIC.()}$ (which we did not pursue further), one could scale each argument-channel of the data to unit amplitude, arriving at the 'scaled' objective function 

$$\begin{equation}\tag{10}
\Phi_{\mathrm{prescaled}} := \Phi\circ\rho \qquad\text{with}\quad \rho(Y):=\mathrm{diag}[\|Y^1\|_\infty^{-1}, \cdots, \|Y^d\|_\infty^{-1}]\cdot (Y^1, \cdots, Y^d)^T.
\end{equation}$$

This version of \texttt{SigIC}, however, is sensitive to outliers in $Y$, which we account for by preceding $\Phi_{\mathrm{prescaled}}$ with an outlier-removal, that is we set

$$\begin{equation}\tag{11}
\Phi_{\mathrm{prepro}} := \Phi_{\mathrm{prescaled}}\circ\varsigma \qquad\text{with}\quad \varsigma \ \text{ freeing $Y$ of outliers up to a given proportion.}
\end{equation}$$

The above auxiliary functions $\rho$ and $\varsigma$ are implemented below.

```python
# Scaling each data-channel to unit amplitude:
def rescale(X):
    '''input: X [pytorch tensor of shape (N,d)]
       returns: rescaled tensor X* = (X[:,i]/(|X[:,i]|.max()) ; i=0,...,d-1)'''
    Y = torch.clone(X)
    for i in range(X.shape[1]):
        Y[:,i] = (1/torch.max(torch.abs(Y[:,i])))*Y[:,i]
    
    return Y

#################################################################################
#################################################################################

def remove_outliers(X, gam):
    Y  = torch.clone(X)
    d  = Y.unsqueeze(1).shape[-1]
    m  = torch.mean(Y,0)
    ym = torch.abs(torch.max(Y - m,0)[0])
    for i in range(d):
        Y  = Y[torch.abs(Y[:,i] - m[i].item()) < (1-gam[i])*ym[i]]
    return Y
```

Implement $\Phi_{\mathrm{prepro}}=:\texttt{SigCF}$ (defined in (11)):

```python
def SigCFrescale(X,mu,winlen, gamma):
    '''input: data X [pytorch tensor of shape (N,d)]; windowlength of sample-ts [winlen; int] (cf. (8));
              'control-tuple' mu [np-array of shape (d-1,), entries ordered by number of channel],
              'outlier-proportion-tuple' gamma [np-array of shape (d,), gamma[i]=proportion of largest values to delete from X[:,i]] 
       returns: Phi_propro(X) (preprocessed SigIC, as defined in (11)) [double] 
    ''' 
    # Preprocess the data:
    Y = torch.clone(X)
    Y = crop_outliers(Y, gamma) #delete outliers (delete the largest gamma[i]*N values from X[:,i])
    Y = torch.from_numpy(Y)     #convert back to torch tensor
    Y = rescale(Y)              #rescale 'outlier-cleaned' data
    
    # Apply SigIC to preprocessed data:
    s = SigIC(Y,mu,winlen)
    
    return s    
```

```python
# Scaling each data-channel to unit amplitude:
def barSigCF(X,mu,winlen):
    '''input: data X [pytorch tensor of shape (N,d)], windowlength of sample-ts [winlen; int] (cf. (8)),
              'control-tuple' mu [np-array of shape (d-1,); entries ordered by number of channel] 
       returns: \bar{Phi}_mu(X) (sum over mu-controlled subselection of normalized cross-cumulants (13)) [double] 
    '''
    d      = X.shape[1] #number of components of X
    
    # Create lists W={W_k, k=2,...,d} of cross-polynomials (cf. (6)/(4)) whose assoc. cumulants (7)/(6) we want to compute:
    W      = [[]]*(d-1)
    mu_max = np.zeros(d-1) #allocate to determine length of longest monomial in W
    
    for k in range(d-1):
        u,v       = ret_uv(k+2)                   #set u=u_k = '12...(k-1)', v=v_k = 'k' for k=2,...,d
        for j in range(int(round(mu[k]))):
            W[k] = W[k] + crosscumu(u,v,j+1)      #assemble W_k(u_k,v_k) = \sqcup_{j=1,...,mu[k]} crosscumu(u,v,j) (cf. (4))
        mu_max[k] = max([deg_poly(q) for q in W[k]]) #determine length of longest monomial in W[k]
        
    mu_max = int(round(np.max(mu_max))) #length of longest monomial in W
    
    ### Print W if you want to see over which cross-shuffled elements q the sum (6) is taken:
    print('W =', W) 
    
    # Compute log-coordinates of ESig(X) up to depth mu_max
    logESX = logESig(X,mu_max,winlen)
    
    s      = np.zeros(d-1)
    # Compute the kth-summand of the outer sum in (13):
    for k in range(d-1):
        # Compute normalization (12) for each q in W[k]:
        norms = dict.fromkeys([tuple(q) for q in W[k]]) #allocate dictionary (with q \in W[k] as keys, i.e.: norms = {q: None}) 
        for q in W[k]:
            ordrep          = ''.join(sorted(q[0])) #returns ordered representative \vec{q} of q
            normfactors_0   = [''.join(g) for _, g in groupby(ordrep)] #select 'monoletteral subwords' of \vec{q}
            # Compute factors according to (12):
            normfactors_1   = [coeffextract(logESX,f,d) if len(f)>1 else np.sqrt(np.abs(coeffextract(logESX,f+f,d))) for f in normfactors_0]
            norms[tuple(q)] = np.prod(normfactors_1) #compute normalisation constant (~ denominator in (12))
            print('norms(q) =',norms[tuple(q)])
            print('coeffs(q) =',coeffextract(logESX,q,d)**2)
        s[k] = sum([(norms[tuple(q)]**(-2))*coeffextract(logESX,q,d)**2 for q in W[k]]) #k-th inner sum in (13)
    
    # Compute \Phi_mu(X) = \sum_{k=0}^{d-2} s[k] (cf. (6))
    SigICX  = np.sum(s)
    
    return SigICX
```

```python
# Auxiliary functions for outlier detection:
def outlier_count(X,alpha):
    '''input: X [(N,)-pytorch tensor], alpha [double between 0 and 1]
       returns: number of points in X lying outside [-alpha, alpha]'''
    
    N      = X.shape[0]
    Y      = (X - torch.mean(X)).numpy()
    Y_abs  = np.abs(Y)
    indout = (Y_abs - alpha > 0).astype(int) #j -> 0 if |X[j]| > alpha (outlier), j -> 1 else [Booleans cast as 0,1]
    
    return sum(indout) #percentage of '>alpha-outliers'

def outlier_percent(X,alpha):
    N = X.shape[0]
    return outlier_count(X,alpha)/N

def outlier_remove(X,alpha):
    '''input: X [(N,d)-pytorch tensor], alpha [(d,)-array]
    '''
    
    [N,d]   = X.shape
    Y       = (X-torch.mean(X,0)).numpy()
    indouts = np.zeros(N)
    
    for i in range(d):
        out_channel = (np.abs(Y[:,i]) - alpha[i] > 0).astype(int)
        out_channel = np.asarray(out_channel)
        #print('out_channel:',sum(out_channel))
        indouts = indouts + out_channel
    
    indouts    = (indouts > 0).astype(int) 
    n_indouts  = sum(indouts)
    #print(n_indouts)
    for j in range(N):
        indouts[j] = j*indouts[j] - (1-indouts[j]) #j if indouts[j]=1 (--> outlier), -1 else
        
    indouts = [j for j in indouts if j>=0] #index-position of outliers
    X_new   = np.zeros((N-n_indouts,d))
    for i in range(d):
        X_new[:,i] = np.delete(X[:,i], indouts)
    
    return X_new

def crop_outliers(X,gamma):
    '''input: X [(N,d)-pytorch tensor], outlier propostions gamma [(d,)-array]'''
    
    d     = X.unsqueeze(1).shape[-1] # get number of channels
    alpha = np.zeros(len(gamma))
    props = np.zeros(len(gamma))
    scale = 0.9
    for i in range(d):
        Y        = X[:,i] - torch.mean(X[:,i])
        xmax     = torch.max(Y).item()
        props[i] = outlier_percent(Y,xmax)
        while props[i] < gamma[i]:
            xmax     = scale*xmax
            props[i] = outlier_percent(Y,xmax)
        alpha[i]     = xmax
    
    return outlier_remove(X,alpha)
```
